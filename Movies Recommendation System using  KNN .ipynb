```python
import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("ratings.csv")
ratings
```

```python
cols = ["movieId","title"]
movies = pd.read_csv("movies.csv", usecols= cols)
movies
```

```python
df= pd.merge(movies, ratings, on= "movieId")
df.drop(columns=["timestamp"], inplace= True)
```

```python
 df
```

```python
#Create UserID Matrix with rating as values
matrix = pd.pivot_table(df, values="rating", index="movieId", columns="userId").fillna(0)
matrix
```

```python
from scipy.sparse import csr_matrix
```

Uing csr_matrix , convert the above matrix into sparses matrix

**Advantages of the CSR format**
- efficient arithmetic operations CSR + CSR, CSR * CSR, etc.
- efficient row slicing
- fast matrix vector products

```python
matrix = csr_matrix(matrix.values)
matrix
```

#### Implement Unsupervised Machine Learning Algorith Using NearestNeighbors

```python
from sklearn.neighbors import NearestNeighbors
#metric= cosine, will look for cosine similarity
# we will recommend movies based on the cosine similarit score
neigh = NearestNeighbors(n_neighbors= 20, algorithm= "brute", metric= "cosine") 
neigh.fit(matrix)
```

```python
#process module from fuzzwuzzy library will speedup the searching process
from fuzzywuzzy import process
```

```python
# it will extract the movie title, the percentage of matching and its index in the dataframe movies
process.extractOne("jumangi", movies["title"]) 
```

neigh.kneighbors returns the distance and indices
- distance : Array representing the lengths to points i.e the cosine distance to the point, only present if return_distance=True, if distance is less, it is highly similiar.
- indices :Indices of the nearest points in the matrix

```python
def movie_recommender(movie_name, matrix, n):
    #idx is the index of the movie_name
    idx = process.extractOne(movie_name, movies["title"])[2]
    print("The movie name is  {} with a index of {}".format(movies["title"][idx], idx))
    print("Recommendaions of {}:".format(movies["title"][idx]))
    distance, indices = neigh.kneighbors(X= matrix[idx],n_neighbors= n+1) #n+1-->since the first movie would be itself
    
    for i in indices:
        print(movies["title"][i].where (i!=idx)) #dropping the movie itself
    
    
```

```python
movie_recommender("jumanji", matrix, 5)
```
